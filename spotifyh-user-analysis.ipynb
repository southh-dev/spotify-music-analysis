# Analysis of my personal music taste, preferences and habits

## The Aim

This is my Personal Project where I'm going to be using spotipy to analyse user data about myself. I intend to analyse my personal music taste and see if i can find any patterns in what i listen to.

The first thing I'm going to do is log in and authenticate myself as the user, this will allow me to actually fetch and retrieve relevant data.

```python
# I'm importing spotipy so i can authenticate myself and retrieve data as necessary
import spotipy
from spotipy.oauth2 import SpotifyOAuth

import config  # Here I'm importing my config.py file, which is explained in the readme.txt

scope = "user-library-read playlist-read-private"

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id=config.CLIENT_ID,
                                               client_secret=config.CLIENT_SECRET,
                                               redirect_uri=config.REDIRECT_URI,
                                               scope=scope))

user = sp.current_user()
print(f"Logged in as: {user['display_name']}")
```

After successfully logging in I'm going to begin by obtaining my top 25 songs from 3 different listening periods. 

### Short Term Listening Habits

Below i'm going to have some data that will show my most listened to songs in the short term, so the last month or so.

```python
top_tracks = sp.current_user_top_tracks(limit=25, time_range='short_term')

import pandas as pd

items = top_tracks['items']

track_data = []

for track in items:
    track_info = {
        'track_name': track['name'],
        'artist_names': ', '.join([artist['name'] for artist in track['artists']]),
        'album_name': track['album']['name'],
        'duration_ms': track['duration_ms'],
        'release_date': track['album']['release_date'],
    }
    track_data.append(track_info)

df_tracks = pd.DataFrame(track_data)

df_tracks['duration_sec'] = df_tracks['duration_ms'] / 1000
df_tracks['duration_min'] = df_tracks['duration_sec'] / 60

df_tracks
```
